```python
b = [20, 10, 5]
total = 0
for e in b:
    total = total + e
print(total)
```

*To ever stop the loop, or "break" it, we can use the "if x ==5: , break" within the while loop as seen below*

```python
x = 0

while x<20:
    # if x ==5:
       #  break
    print(x)
    x += 1 # This part here just means you're iterating x by blocks of (1) or whatever multiple u want
```

```python
x = 0

while True:
    if x ==5:
        break
    print(x)
    x+=1 
```

# From honours code

```python
import statsmodels.regression.linear_model as lm
import statsmodels.formula.api as sm
import statsmodels.tools.tools as ct
from sklearn.linear_model import LinearRegression
import statsmodels.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing CSV Files

```python
data = pd.read_csv('clean_returns.csv', index_col = 0)

factor = pd.read_csv('factor5.csv', index_col = 0)
factor = smf.add_constant(factor)

factor.head()
```

### SINGLE-PERIOD FACTOR ATTRIBUTION CODE ###

```python
final = []
pvalues = []
rsq =[]

funds = len(data.columns)

for x in range(0,funds):
    cat = data.iloc[:,0+x] # the "0 + x" here is key, making sure that every iteration of 'cat' is a new column (the rows are ALL select) ///// add int(x) if not working
    
    results = smf.OLS(cat, factor).fit() # Fitting the funds to the factors with an OLS model
    
    final.append(results.params) #Appending the coefficients to the empty list that had been created
    pvalues.append(results.pvalues)
    rsq.append(results.rsquared) 
    
final = pd.DataFrame(final, index = data.columns)
pvalues = pd.DataFrame(pvalues, index = data.columns)
rsq = pd.DataFrame(rsq, index = data.columns)
```

Finding the best ranking stocks by sorting by highest Alpha

```python
final.sort_values(by = ['const'], ascending = False)
```

Joining coefficients with the R-squared

```python
final2 = pd.concat([final, rsq], axis=1)
final2
```

‘can the model predict if a billionaire gained their wealth from founding the company (not inheriting or acquiring)?’ - https://jkaterina.com/billionaires-regression/#data-cleaning

```python
import os 
```

```python
path = os.getcwd() 

apple = pd.read_csv('billionaires.csv')
apple.head(4)

#For this exercise, we are not interested in the time dimension, rather the first entry for each unit - the year when they were first ranked billionaire (DROP REST)
```

<br>

**"ignore_index = True"** ////  Pandas will ignore the index values in the inputs, and will generate a new index for the output<br><br> != means not equal

```python
# drop duplicates based on the name column

apple = apple.drop_duplicates(subset = ['name'], ignore_index =True)
apple.head(5) 
```

<br>

Cleaning By re-defining apple progressively based on values we don't want within columns

```python
apple= apple[(apple['demographics.age'] > 0)]
apple = apple[(apple['company.founded'] != 0)]
apple = apple[(apple['location.region'] != '0')] # x = x[(x['chicken_column'] > 40)] //// Within x you're only selecting the column chicken_column for values above 40.
apple = apple[(apple['wealth.how.category'] != '0')]
```

```python
apple['demographics.age'].min() # checking
```

After that, the variable company age was created, where ‘company age’ refers to the age of the company when its owner was ranked as a billionaire. Also, the variable ‘gender’ was renamed to match the integer data type.

```python
#age of company when the founder became billionaire
apple['company age'] = apple['year'] - apple['company.founded']
```

Essentially here below, we are changing male and female into boolean

```python
apple.loc[apple['demographics.gender'] == 'male', 'demographics.gender'] = 1
apple.loc[apple['demographics.gender'] == 'female', 'demographics.gender'] = 0
```
